# Stock Sentimental Analysis using News Headlines

```python
# Import pandas library 
import pandas as pd
```

```python
# Extract database with encode techniques because databse contains utf characters
df = pd.read_csv('Data.csv',encoding= "ISO-8859-1")
```

```python
df.head()
```

```python
# Lets split the data into train and test to avoid a data leakage
train = df[df['Date'] < '20150101']
test = df[df['Date']> '20141231']
```

```python
# Removing punctuations
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
```

```python
# Renaming column names for easy access
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index
data.head(5)
```

```python
# Convertng all headlines into lower case
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
' '.join(str(x) for x in data.iloc[0,0:25])
```

```python
# Join all 25 columns into one column
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[3]
```

```python
## Join all 25 columns into one column this for test data
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
```

```python
test_transform[0]
```

# Vectorization

```python
# Importing required library for vectorization and for ML model building
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
## Implementing BAG OF WORDS
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

# RandomForest

```python
# Implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
test_dataset = countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```

# **MultinomialNB**

```python
# Importing multinomial naive_bayes library and perform the fit the data
from sklearn.naive_bayes import MultinomialNB
gnb = MultinomialNB()
gnb.fit(traindataset,train['Label'])
```

```python
# Predictions for new data using multinomial naive_bayes Ml Model
predictions = gnb.predict(test_dataset)
```

```python
# Check accuracy of this model
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```

# XGboost

```python
# Importing XGboost library and perform the fit the data
from xgboost import XGBClassifier
xg = XGBClassifier()
xg.fit(traindataset,train['Label'])
```

```python
# Predictions for new data using Xgboost Ml Model
predictions = xg.predict(test_dataset)
```

```python
# Check accuracy of this model
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
